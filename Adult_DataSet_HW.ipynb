```python
#ip/op Creation 
ip = adult.drop('AnnualIncome',axis=1)
op = adult.AnnualIncome
```

```python
ip.head()
```

```python
op.head()
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(ip,op,test_size=0.2,random_state=5)
```

```python
x_train #Randomly takes data for training ... Agar dekha hei toh upar se phir se run karo dubara run karo....
```

```python
#Standard Scaler Transform 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_train = sc.fit_transform(x_train) #fit matlab formula apply karna aur sikhna #transform matlab formula apply karne ke baad jo result return karega...
x_test = sc.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
pred = lr.predict(x_test)
pred
```

```python
y_test
```

```python
#Accuracy
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
ac = accuracy_score(y_test,pred)
re = recall_score(y_test,pred)
pr = precision_score(y_test,pred)
f1 = f1_score(y_test,pred)

print('Accuracy : ',ac)
print('Recall : ',re)
print('Precision : ',pr)
print('F1 Score : ',f1)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test,pred)
cm
```

```python
cmd = ConfusionMatrixDisplay(cm)
cmd.plot()
plt.show()
```
