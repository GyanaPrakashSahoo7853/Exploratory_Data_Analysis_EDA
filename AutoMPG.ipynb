```python
#ip/op Creation 
ip = auto.drop('MPG',axis=1)
op = auto.MPG
```

```python
ip.head()
```

```python
op.head()
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(ip,op,test_size=0.2,random_state=5)
```

```python
x_train
```

```python
#Standard Scaler Transform 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_train = sc.fit_transform(x_train) 
x_test = sc.fit_transform(x_test)
```

```python
#Linear Regression 
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
pred = lr.predict(x_test)
pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test,pred)
r2 = r2_score(y_test,pred)

print("MSE :" , mse ) #error jitna kam model utna accurate hei #it just shows the error..
print("R2 Score: ",int(r2*100),'%')
```
